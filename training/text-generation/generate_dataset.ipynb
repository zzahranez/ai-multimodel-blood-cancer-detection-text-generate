```python
import pandas as pd
import os
```

```python

import glob

# 1. Tentukan lokasi folder data
data_dir = r'D:\Lab Experiment\Deep Learning\blood cancer detection and text generate\training\text-generation\data'
output_file = 'dataset_final.csv'

all_entries = []

txt_files = glob.glob(os.path.join(data_dir, "*.txt"))

for file_path in txt_files:

    label = os.path.basename(file_path).replace(".txt", "")
    
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        
        for line in lines:
            clean_text = line.strip()
            
            if clean_text:
     
                if not clean_text.endswith("<|endoftext|>"):
                    clean_text += "<|endoftext|>"
                
                formatted_line = f"### Jenis Sel: {label} ### Penjelasan: {clean_text}"
                all_entries.append(formatted_line)

# 3. Masukkan ke Pandas DataFrame
df = pd.DataFrame(all_entries, columns=['text'])

# 4. Quality Control: Hapus duplikat agar model tidak overfitting
initial_count = len(df)
df = df.drop_duplicates()
final_count = len(df)

# 5. Simpan ke CSV
df.to_csv(output_file, index=False, encoding='utf-8')

print(f"--- Selesai Menjahit Dataset ---")
print(f"Total baris awal: {initial_count}")
print(f"Total baris unik (setelah cleaning): {final_count}")
print(f"File disimpan di: {output_file}")
```
