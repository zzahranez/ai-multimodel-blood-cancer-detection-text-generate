```python
import pandas as pd
from transformers import AutoTokenizer
```

```python
df = pd.read_csv("dataset_final_gpt2.csv")
print(df.head())
```

```python
tokenizer = AutoTokenizer.from_pretrained("flax-community/gpt2-small-indonesian")

def count_tokens(text):
    tokens = tokenizer.encode(text)
    return len(tokens)

# 3. Hitung panjang tiap baris
df['token_length'] = df['text'].apply(count_tokens)

# 4. Tampilkan Statistiknya
print(f"--- Statistik Dataset Zahran ---")
print(f"Panjang Token Terpendek : {df['token_length'].min()}")
print(f"Panjang Token Rata-rata : {df['token_length'].mean():.2f}")
print(f"Panjang Token Terpanjang: {df['token_length'].max()}") # INI YANG PENTING!

# 5. Cek baris mana yang paling panjang (biar bisa kamu intip)
longest_row = df.loc[df['token_length'].idxmax(), 'text']
print(f"\nContoh baris terpanjang:\n{longest_row[:150]}...")
```
