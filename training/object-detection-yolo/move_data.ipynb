```python
import pandas as pd
import torch
import os
import shutil
```

```python
def Show(dataset):
    for i in dataset:
        print(i)
```

```python
base_dir = r'D:/Lab Experiment/Deep Learning/Blood Cancer Object Detection'
source_dir = r'D:/Lab Experiment/Deep Learning/Blood Cancer Object Detection/archive'
dir_in_archive = os.listdir(source_dir)
```

```python
Show(dir_in_archive)
```

## **Make Directory Base On Data Class**

```python
if not os.path.exists("data"):
    os.mkdir("data")
else:
    print("data directory already exists,")
```

```python

for dir_class in dir_in_archive:
    dataset_dir = f"{base_dir}/data"
    print(dataset_dir)
    dataset_path = os.path.join(dataset_dir, dir_class)
    if not os.path.exists(dataset_path):
        os.mkdir(dataset_path)
    else:
        print(f"{dataset_path} already exists,")
```

## **Make Directory For All Class**

### **a. For Resource**

```python
dataset_dir = []
for i in dir_in_archive:
    dataset_path = os.path.join(source_dir, i)
    dataset_dir.append(dataset_path)
```

```python
Show(dataset_dir)
```

### **b. for destionation**

```python
destination_dir = []
for ds in dir_in_archive:
    path = f"{base_dir}/data/{ds}"
    destination_dir.append(path)
```

```python
Show(destination_dir)
```

### Copy Data

```python
for idx, class_dir in enumerate(dataset_dir):
    print(f"ini adalah class ke -{idx} dengan nama {class_dir}")
    for img in os.listdir(class_dir)[:500]:
        img_source_path = os.path.join(class_dir, img)
        shutil.copy(img_source_path, destination_dir[idx])
        
```
