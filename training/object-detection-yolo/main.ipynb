#### **a. Declarate Basic funciton for show image and initialize the dir variabel**

```python
# Show Image
def Show_Image(img_dir):
    print("Jumlah gambar: ", len(os.listdir(img_dir)))
    fig, axes = plt.subplots(1, 3, figsize=(6, 3))
    for idx, i in enumerate(os.listdir(img_dir)[:3]):
        img_path = os.path.join(img_dir, i)
        image = Image.open(img_path)
        img_name = i.split("_")[:-1]

        image = Image.open(img_path)
        axes[idx].imshow(image)
        axes[idx].set_title(i.split(".")[0])
        axes[idx].axis('off') 

    plt.tight_layout()  
    plt.show()
```

```python
base_dir = r"D:\Lab Experiment\Deep Learning\blood cancer detection and text generate\training\object-detection-yolo"
val_dir = os.path.join(base_dir,r"dataset\valid\images")
train_dir = os.path.join(base_dir,r"dataset\train\images")
test_dir = os.path.join(base_dir,r"dataset\test\images")
```

```python
labels_dir = r"dataset\train\labels"
counter = Counter()

for file in os.listdir(labels_dir):
    if file.endswith(".txt"):
        with open(os.path.join(labels_dir, file)) as f:
            for line in f:
                cls = int(line.split()[0])
                counter[cls] += 1
                
print(counter)
```

### **2. Load YOLO Architecture**

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
print(model)
```

### **3. Training Data**

```python
result = model.train(
    data = "data.yaml",
    epochs = 100,
    batch = 16,
    imgsz=640,
    device =0,
    seed=404,
    workers=4,
    half=True,
    augment=True,
    optimizer="AdamW",
    lr0=0.005,
    lrf=0.01,
    momentum=0.937,
    weight_decay=0.0005,
)
```
